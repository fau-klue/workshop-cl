```python
import gzip
lemmata = list()
with open('../data/cl-stanza.conllu', 'rt') as f:
    for line in f:
        line = line.rstrip()
        if not line.startswith("#") and line != "":
            lemmata.append(line.split("\t")[2].lower())

len(lemmata)
```

```python
from collections import Counter
freq = Counter(lemmata)
```

```python
from pandas import DataFrame
df_freq = DataFrame.from_dict(freq, orient='index', columns=['f1']).sort_values(by='f1', ascending=False)
df_freq
```

```python
df_freq['N1'] = df_freq['f1'].sum()
df_freq
```

```python
from pandas import read_csv
df_ref = read_csv("../data/bnc-lemma-lower-freq.tsv.gz",
                  dtype=str, sep="\t", quoting=3, keep_default_na=False, index_col=1, names=["f2"])
df_ref["f2"] = df_ref["f2"].astype(int)
```

```python
df_ref.loc[""]
N2 = int(df_ref.loc[""]["f2"])
df_ref = df_ref.drop("")
df_ref
```

```python
df_ref['N2'] = N2
df_ref
```

```python
df = df_freq.join(df_ref).fillna(0, downcast='infer')
```

```python
from association_measures.measures import score
df_scored = score(df)
```
